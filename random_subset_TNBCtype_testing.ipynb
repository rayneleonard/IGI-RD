```python
import pandas as pd
import numpy as np

my_file = "master_ref_set.csv" #insert file name here, should be in same directory if no file path
df = pd.read_csv(my_file)

# Create index and sample dataframes
index_df = df.drop(df.columns[2:], axis=1) 
sample_df = df.drop(df.columns[0:2], axis=1)

countif = 0

      
for i in range(1,101): #can change 101 to whatever number of output files you need
    countif = str(i)
    outfilename = countif + "_result.csv" # can change _result to whatever file name you desire
    random_df = sample_df.sample(n=10, axis=1) # n=10 denotes how many random samples to pull
    res_df = pd.concat([index_df,random_df], axis=1)
    res_df.to_csv(outfilename, index=True)
    #res_df.to_csv('random_"countif".csv', index=True)
   

res_df.head() #displays below, so you know this worked
```
